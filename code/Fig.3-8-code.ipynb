```python
import os
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import re
import matplotlib as mpl
from scipy.stats import gaussian_kde
plt.rcParams['xtick.direction'] = 'in'
plt.rcParams['ytick.direction'] = 'in'
plt.rcParams['font.family'] ='serif'

label_to_rgb = {
    '0': [31, 120 ,180],
    '1': [174 ,199, 231],
    '2': [255, 127,  14],
    '3': [255 ,187, 120],
    '4': [ 44, 160,  44],
    '5': [152 ,223, 138],
    '6': [214 , 39,  39],
    '7': [255, 152, 150],
    '8': [148, 103 ,189],
    '9': [197 ,176 ,213],
    '10': [140 , 86  ,75],
    '11': [196, 156 ,148],
    '12': [227, 119, 194],
    '13': [247 ,182 ,210],
    '14': [127, 127, 127],
    '15': [199 ,199, 199],
    '16': [188 ,188,  34],
    '17': [219, 219 ,141],
    '18': [ 23 ,190, 207],
    '19': [158 ,218, 229],
}


from tqdm import tqdm

path = os.getcwd()
files = os.listdir(path)
files = [f for f in files if f.endswith('.txt')]
sorted_lst = sorted(files, key=lambda x: int(x[:-4]))



datas = []
for file in sorted_lst:
    data = pd.read_csv(file, sep='\t')
    datas.append(data)



fig, axs = plt.subplots(5, 4, figsize=(15, 17),dpi = 1000)
plt.rcParams.update({'font.family': 'serif', 'font.size': 18})


for row in tqdm(range(5)):
    for col in tqdm(range(4)):
        all_pc1 = []
        all_pc2 = []
        indices = row * 4 + col
        for i in range(indices+1):
            pc1 = datas[i]['PC1']
            pc2 = datas[i]['PC2']
            all_pc1.extend(pc1)
            all_pc2.extend(pc2)
            # axs[row, col].scatter(datas300[i]['PC1'].values, datas300[i]['PC2'].values, c=np.array(label_to_rgb[str(i)])/255,s=1)
            # axs[row, col].scatter(datas1500[i]['PC1'].values, datas1500[i]['PC2'].values, c=np.array(label_to_rgb[str(i)])/255,s=1)
            axs[row, col].scatter(datas[i]['PC1'].values, datas[i]['PC2'].values, c=np.array(label_to_rgb[str(i)])/255,s=1)

        data = np.vstack([all_pc1, all_pc2])
        kde = gaussian_kde(data)
        xgrid = np.linspace(-90, 90, 100)
        ygrid = np.linspace(-90, 90, 100)
        X, Y = np.meshgrid(xgrid, ygrid)
        Z = kde(np.vstack([X.ravel(), Y.ravel()]))
        Z = Z.reshape(X.shape)

        # Plot KDE shading under scatter
        axs[row, col].contourf(X, Y, Z, cmap='Blues', alpha=0.3)

        # Set axis limits and ticks
        axs[row, col].set_xlim((-90, 90))
        axs[row, col].set_xticks([-50, 0, 50])
        axs[row, col].set_xticklabels([-50, 0, 50], fontweight='bold',fontsize=18)  # 加粗 x 轴刻度标签
        axs[row, col].set_ylim((-90, 90))
        axs[row, col].set_yticks([-50, 0, 50])
        axs[row, col].set_yticklabels([-50, 0, 50], fontweight='bold',fontsize=18)  # 加粗 y 轴刻度标签


plt.tight_layout()  # Adjust layout


     

```

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import gaussian_kde

# Set global font and font size
plt.rcParams.update({'font.family': 'serif', 'font.size': 18})

# Define the output directory
output_dir = os.getcwd()

# Create a figure and axes
fig, axs = plt.subplots(2, 2, figsize=(8, 8), dpi=1000)

# Define categories
categories = {
    " C - N rings": [0, 1, 2, 13, 14],
    "Phenyl rings": [3, 7, 9, 18, 19],
    "Chain - like structures": [12, 11, 16, 5],
    "Ionic salts": [4, 6, 10]
}

# Flatten the axes for easy iteration
axs = axs.flatten()

# Iterate over categories and plot data
for i, (category, indices) in enumerate(categories.items()):
    all_pc1 = []
    all_pc2 = []
    
    for index in indices:
        pc1 = datas[index]['PC1']
        pc2 = datas[index]['PC2']
        all_pc1.extend(pc1)
        all_pc2.extend(pc2)
        axs[i].scatter(pc1, pc2, c=np.array(label_to_rgb[str(index)]) / 255, s=5)

    # KDE shading
    data = np.vstack([all_pc1, all_pc2])
    if len(data[0]) > 1:  # Ensure KDE can be computed
        kde = gaussian_kde(data)
        xgrid = np.linspace(-90, 90, 100)
        ygrid = np.linspace(-90, 90, 100)
        X, Y = np.meshgrid(xgrid, ygrid)
        Z = kde(np.vstack([X.ravel(), Y.ravel()]))
        Z = Z.reshape(X.shape)

        axs[i].contourf(X, Y, Z, cmap='Blues', alpha=0.3)

    # Set axis limits and ticks
    axs[i].set_xlim((-90, 90))
    axs[i].set_xticks([-50, 0, 50])
    axs[i].set_ylim((-90, 90))
    axs[i].set_yticks([-50, 0, 50])
    axs[i].set_title(category)

# Adjust layout and save the figure
plt.tight_layout()
fig.savefig(f'{output_dir}scatter_with_kde_2x2.png')

plt.show()
```
